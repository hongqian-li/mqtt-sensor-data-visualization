```python
#Subscriber
import paho.mqtt.client as mqtt
import time, ssl

broker = "test.mosquitto.org" #test.mosquitto.org #opendata.technikum-wien.at
topic = "/fhtw/appcsqtt" #"/fhtw/appcs" #"/fhtw/3928ud89/2309irm/2o9di2"

# The callback for when the client receives a CONNACK response from the server.
def on_connect(client, userdata, flags, reason_code, properties):
    # Subscribing in on_connect() means that if we lose the connection and
    # reconnect then subscriptions will be renewed.
    client.subscribe(topic)
    print(f"{reason_code}! We are connected to topic: {topic}")

# The callback for when a PUBLISH message is received from the server.
def on_message(client, userdata, msg):
    print("received message:", str(msg.payload.decode("utf-8")))

client = mqtt.Client(mqtt.CallbackAPIVersion.VERSION2)
client.on_connect = on_connect
client.on_message = on_message

# Configure TLS settings to use port 8883 otherwise use 1883
client.tls_set(cert_reqs = ssl.CERT_NONE)
client.tls_insecure_set(True)

print("connecting to broker ...")
client.connect(broker, 8883, 60)

#client.loop_forever()
client.loop_start()
time.sleep(20)
client.loop_stop()
client.disconnect()
print("END")
```

```python
#Publisher
import paho.mqtt.client as mqtt
import time, ssl

broker = "test.mosquitto.org" #test.mosquitto.org #opendata.technikum-wien.at
topic = "/fhtw/appcsqtt" #"/fhtw/appcs" #"/fhtw/3928ud89/2309irm/2o9di2"

client = mqtt.Client(mqtt.CallbackAPIVersion.VERSION2)

# Configure TLS settings to use port 8883 otherwise use 1883
client.tls_set(cert_reqs = ssl.CERT_NONE)
client.tls_insecure_set(True)

print("connecting to broker ...")
client.connect(broker, 8883, 60)

payload = "hello students" #payload must be a string, byte-array, int or float e.g. payload = random.randint(0,9)
client.publish(topic, payload)
print("data published")
client.disconnect()
print("END")
```
