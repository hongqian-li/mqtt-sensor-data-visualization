### JSON in Python
Receiving data as a String and parsing it into a Dictionary

```python
import json

message = '{"u1":235,"power":4080,"freq":50.1}' #to simulate received data via MQTT
print(message, type(message))

jsondata = json.loads(message)
print(jsondata, type(jsondata))
```

Creating a Dictionary and parsing it into a String (to be sent via MQTT later)

```python
#Fictious sensor values as integer
voltage = 234
powervalue = 4090
frequency = 49.98

#dictionary containing the data to be sent
data = {
    "u1": voltage,
    "power": powervalue,
    "freq": frequency
}

print(data, type(data)) #data is a dictionary

#dump dictionary data to get a string (needed for MQTT)
payload = json.dumps(data)

#payload is now ready to be sent via MQTT
print(payload, type(payload))
```

### UNIX Timestamp (optional)

```python
from datetime import datetime

now = datetime.now() #current timestamp
print("now        ", now)

timestamp = datetime.timestamp(now)
print("timestamp: ", timestamp)

#replace the dot by nothing and cut the last 3 characters to get the timestamp in milliseconds
ts = str(timestamp).replace(".","")[:-3]
ts = int(ts) #parse from string to int for further use
print("ts:        ", ts)
```

```python
#Fictious sensor values as integer
voltage = 234
powervalue = 4090
frequency = 49.98

#dictionary containing the data to be sent
data = {
    "u1": voltage,
    "power": powervalue,
    "freq": frequency,
    "time": ts
}

print(data, type(data)) #data is a dictionary

#dump dictionary data to get a string (needed for MQTT)
payload = json.dumps(data)

#payload is now ready to be sent via MQTT
print(payload, type(payload))
```
